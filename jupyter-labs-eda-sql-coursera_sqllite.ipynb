### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to get unique launch sites
query = "SELECT DISTINCT launch_site FROM SPACEXTBL;"

# Execute the query and fetch the results
cur.execute(query)
unique_launch_sites = cur.fetchall()

# Print the unique launch sites
for site in unique_launch_sites:
    print(site[0])

# Close the connection
con.close()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to get 5 records where launch sites begin with 'CCA'
query = "SELECT * FROM SPACEXTBL WHERE launch_site LIKE 'CCA%' LIMIT 5;"

# Execute the query and fetch the results
cur.execute(query)
records = cur.fetchall()

# Print the records
for record in records:
    print(record)

# Close the connection
con.close()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to get the total payload mass carried by boosters launched by NASA (CRS)
query = """
SELECT SUM(PAYLOAD_MASS__KG_) AS total_payload_mass
FROM SPACEXTBL
WHERE Customer == 'NASA (CRS)';
"""

# Execute the query and fetch the result
cur.execute(query)
result = cur.fetchone()

# Print the total payload mass
total_payload_mass = result[0] if result[0] is not None else 0
print("Total payload mass carried by boosters launched by NASA (CRS):", total_payload_mass, "kg")

# Close the connection
con.close()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to get the average payload mass carried by booster version F9 v1.1
query = """
SELECT AVG(PAYLOAD_MASS__KG_) AS average_payload_mass
FROM SPACEXTBL
WHERE Booster_Version = 'F9 v1.1';
"""

# Execute the query and fetch the result
cur.execute(query)
result = cur.fetchone()

# Print the average payload mass
average_payload_mass = result[0] if result[0] is not None else 0
print("Average payload mass carried by booster version F9 v1.1:", average_payload_mass, "kg")

# Close the connection
con.close()
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to get the date of the first successful landing outcome in ground pad
query = """
SELECT Date
FROM SPACEXTBL
WHERE Landing_Outcome LIKE '%Success (ground pad)%'
ORDER BY Date ASC
LIMIT 1;
"""

# Execute the query and fetch the result
cur.execute(query)
result = cur.fetchone()

# Print the date of the first successful landing outcome in ground pad
first_successful_landing_date = result[0] if result else "No successful landing found"
print("Date of the first successful landing outcome in ground pad:", first_successful_landing_date)

# Close the connection
con.close()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to list the names of boosters with specific criteria
query = """
SELECT Booster_Version
FROM SPACEXTBL
WHERE Landing_Outcome LIKE '%Success (drone ship)%'
  AND PAYLOAD_MASS__KG_ > 4000
  AND PAYLOAD_MASS__KG_ < 6000;
"""

# Execute the query and fetch the results
cur.execute(query)
results = cur.fetchall()

# Print the names of the boosters
booster_names = [row[0] for row in results]
print("Boosters with successful drone ship landings and payload mass between 4000 and 6000 kg:", booster_names)

# Close the connection
con.close()
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to list the total number of successful and failed mission outcomes
query = """
SELECT Mission_Outcome, COUNT(*) AS total_count
FROM SPACEXTBL
GROUP BY Mission_Outcome;
"""

# Execute the query and fetch the results
cur.execute(query)
results = cur.fetchall()

# Print the total number of successful and failed mission outcomes
for outcome, count in results:
    print(f"Mission Outcome: {outcome}, Total Count: {count}")

# Close the connection
con.close()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# Load the SQL extension
#%load_ext sql

# Connect to SQLite database
#import sqlite3
con = sqlite3.connect("my_data1.db")

# Register the connection with the %sql magic
%sql sqlite:///my_data1.db

# Query to list the required records
%sql SELECT strftime('%m', Date) AS Month, Booster_Version, Launch_Site, Landing_Outcome \
     FROM SPACEXTABLE \
     WHERE Landing_Outcome LIKE '%Failure%' AND Landing_Outcome LIKE '%drone ship%' AND strftime('%Y', Date) = '2015';
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL query to rank the count of landing outcomes between specific dates in descending order
query = """
SELECT Landing_Outcome, COUNT(*) AS outcome_count
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY outcome_count DESC;
"""

# Execute the query and fetch the results
cur.execute(query)
results = cur.fetchall()

# Print the landing outcomes and their counts
for outcome, count in results:
    print(f"Landing Outcome: {outcome}, Count: {count}")

# Close the connection
con.close()
```
