```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def color(value):
    if value == 1:
        return 'green'
    if value == 0:
        return 'red'
    else:
        return 'yellow'

launch_sites_df['marker_color'] = spacex_df['class'].apply(color)
launch_sites_df.head()
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster


# Define initial coordinates for the map
initial_coords = [29.559684, -95.083097]

# Create a folium map centered at the initial coordinates
site_map = folium.Map(location=initial_coords, zoom_start=5)

# Add MarkerCluster to the map
marker_cluster = MarkerCluster().add_to(site_map)

# Add circles for launch sites in launch_sites_df
for _, row in launch_sites_df.iterrows():
    circle = folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=1000,
        color='#0000FF',  # Blue color
        fill=True,
        fill_color='#0000FF'
    ).add_child(folium.Popup(row['Launch Site']))
    site_map.add_child(circle)

# Add markers for spacex_df with customized icons
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=record['Launch Site']
    )
    marker_cluster.add_child(marker)

# Display the map (only works in a Jupyter Notebook environment)
site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from math import radians, sin, cos, sqrt, atan2



# Function to calculate the Haversine distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of the Earth in kilometers

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance

# Coordinates of the coastline
coastline_lat = 28.56443
coastline_long = -80.56222

# Calculate distance to the closest coastline
launch_sites_df['distance_to_coastline'] = launch_sites_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], coastline_lat, coastline_long), axis=1)

# Find the launch site closest to the coastline
closest_site = launch_sites_df.loc[launch_sites_df['distance_to_coastline'].idxmin()]

# Print the closest site and its distance to the coastline
print(f"Closest Launch Site: {closest_site['Launch Site']}")
print(f"Distance to Coastline: {closest_site['distance_to_coastline']} km")

# Define initial coordinates for the map
initial_coords = [29.559684, -95.083097]

# Create a folium map centered at the initial coordinates
site_map = folium.Map(location=initial_coords, zoom_start=5)

# Add MarkerCluster to the map
marker_cluster = MarkerCluster().add_to(site_map)

# Add circles for launch sites in launch_sites_df
for _, row in launch_sites_df.iterrows():
    circle = folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=1000,
        color='#0000FF',  # Blue color
        fill=True,
        fill_color='#0000FF'
    ).add_child(folium.Popup(f"{row['Launch Site']}: {row['distance_to_coastline']:.2f} km"))
    site_map.add_child(circle)

# Add a marker for the closest coastline
folium.Marker(
    location=[coastline_lat, coastline_long],
    icon=folium.Icon(color='blue', icon_color='white'),
    popup='Closest Coastline'
).add_to(site_map)

# Display the map (only works in a Jupyter Notebook environment)
site_map
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from math import radians, sin, cos, sqrt, atan2

# Sample DataFrame (replace this with your actual DataFrame)
data = {
    'Launch Site': ['Site 1', 'Site 2', 'Site 3', 'Site 4'],
    'Lat': [28.562302, 34.632834, 29.559684, 32.640054],
    'Long': [-80.577356, -120.610746, -95.083097, -117.841995]
}
launch_sites_df = pd.DataFrame(data)

# Function to calculate the Haversine distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of the Earth in kilometers

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance

# Coordinates of the coastline
coastline_lat = 28.56443
coastline_long = -80.56222

# Coordinates of other points of interest
railway_lat = 28.56353
railway_lon = -80.58691

highway_lat = 25.56259
highway_lon = -80.57067

city_lat = 28.10106
city_lon = -80.62918

# Calculate distance to the closest coastline
launch_sites_df['distance_to_coastline'] = launch_sites_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], coastline_lat, coastline_long), axis=1)

# Find the launch site closest to the coastline
closest_site_idx = launch_sites_df['distance_to_coastline'].idxmin()
closest_site = launch_sites_df.loc[closest_site_idx]

# Print the closest site and its distance to the coastline
print(f"Closest Launch Site: {closest_site['Launch Site']}")
print(f"Distance to Coastline: {closest_site['distance_to_coastline']:.2f} km")

# Calculate distances to other points of interest
launch_sites_df.loc[closest_site_idx, 'distance_to_railway'] = haversine(closest_site['Lat'], closest_site['Long'], railway_lat, railway_lon)
launch_sites_df.loc[closest_site_idx, 'distance_to_highway'] = haversine(closest_site['Lat'], closest_site['Long'], highway_lat, highway_lon)
launch_sites_df.loc[closest_site_idx, 'distance_to_city'] = haversine(closest_site['Lat'], closest_site['Long'], city_lat, city_lon)

# Print distances to other points of interest
print(f"Distance to Railway: {launch_sites_df.loc[closest_site_idx, 'distance_to_railway']:.2f} km")
print(f"Distance to Highway: {launch_sites_df.loc[closest_site_idx, 'distance_to_highway']:.2f} km")
print(f"Distance to City: {launch_sites_df.loc[closest_site_idx, 'distance_to_city']:.2f} km")

# Define initial coordinates for the map
initial_coords = [29.559684, -95.083097]

# Create a folium map centered at the initial coordinates
site_map = folium.Map(location=initial_coords, zoom_start=5)

# Add MarkerCluster to the map
marker_cluster = MarkerCluster().add_to(site_map)

# Add circles for launch sites in launch_sites_df
for _, row in launch_sites_df.iterrows():
    circle = folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=1000,
        color='#0000FF',  # Blue color
        fill=True,
        fill_color='#0000FF'
    ).add_child(folium.Popup(f"{row['Launch Site']}: {row['distance_to_coastline']:.2f} km"))
    site_map.add_child(circle)

# Add a marker for the closest coastline
folium.Marker(
    location=[coastline_lat, coastline_long],
    icon=folium.Icon(color='blue', icon_color='white'),
    popup='Closest Coastline'
).add_to(site_map)

# Draw a PolyLine between the closest launch site and the coastline
folium.PolyLine(
    locations=[[closest_site['Lat'], closest_site['Long']], [coastline_lat, coastline_long]],
    color='red',
    weight=2.5,
    opacity=1
).add_to(site_map)

# Add markers and lines for other points of interest
points_of_interest = [
    ('Railway', railway_lat, railway_lon, launch_sites_df.loc[closest_site_idx, 'distance_to_railway']),
    ('Highway', highway_lat, highway_lon, launch_sites_df.loc[closest_site_idx, 'distance_to_highway']),
    ('City', city_lat, city_lon, launch_sites_df.loc[closest_site_idx, 'distance_to_city'])
]

for name, lat, lon, distance in points_of_interest:
    folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='green', icon_color='white'),
        popup=f'{name}: {distance:.2f} km'
    ).add_to(site_map)
    
    folium.PolyLine(
        locations=[[closest_site['Lat'], closest_site['Long']], [lat, lon]],
        color='orange',
        weight=2.5,
        opacity=1
    ).add_to(site_map)

# Display the map (only works in a Jupyter Notebook environment)
site_map

# Save the map to an HTML file
#site_map.save('launch_sites_map_with_polyline_and_markers.html')
```
